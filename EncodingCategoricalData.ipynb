# Step # 2: Load the Data into Python

```python
#specify URL where data is located
url = 'https://raw.githubusercontent.com/sanjaygupta1963/Pythoncoding/main/price_train_r_exam_sem2.csv'

# Load Dataset from the Github URL
# pd.read_csv() function: Read a comma-separated values (csv) file into DataFrame.
dframe = pd.read_csv(url)
dframe.info()
```

# Step # 3: Cleaning the Data

```python
# Convert the Entire Dataframe Columns into the Object (String) Datatype
# This is pre-requiste to run the String Repalcement Command in the next Row
# obj_df = obj_df.applymap(str)
# obj_df.info()
```

```python
# Removing All the possible Special Characters from the Entire Dataframe through a FOR Loop
special_char=[';','%','$','!','&','{','}','"','/']
n=0
for i in special_char:
    dframe = dframe.replace(special_char[n],'',regex=True)
    # print(special_char[n])
    n += 1
# For Loop Ends

# Display the first 10 rows of the dataset
dframe.head(10)
```

```python
# Note that Comma ',' should not be repalced from the Amenities Columns , so it has to be replaced through a separate Command.
# Note that Dollar Symbol '$' is not getting replaced by the above For Loop Command, 
# therefore so it has to be replaced through a separate Command.
# Removing the $ sign from the cleaning_fee and price Columns 
dframe['price'] = dframe['price'].str.replace('$','',regex=True)
dframe['price'] = dframe['price'].str.replace(',','',regex=True)
dframe['cleaning_fee'] = dframe['cleaning_fee'].str.replace('$','',regex=True)
dframe['cleaning_fee'] = dframe['cleaning_fee'].str.replace(',','',regex=True)
dframe.head(10)
```

# Method # 1 Using Get Dummies for converting Categorical Labels into Dummy Variables

```python
# Create Dummy Variables for property_type variable with Prefix parameter
# ptype = pd.get_dummies(obj_df['property_type'],prefix="pt")

# In the below code snippet, we passed an argument drop_first=True, 
# this indicates it drops one Dummy Variable from all the dummy variables created for a feature. 
# It is to be dropped because it causes Dummy variable Trap, 
# which is called as some of the features are highly correlated which results in predicting of another feature. 
# So to avoid this situation, we are dropping one dummy variable created after encoding.

# hrtime = pd.get_dummies(dframe['host_response_time'],drop_first=True)


# Create Dummy Variables for host_response_time variable without Prefix parameter
hrtime = pd.get_dummies(dframe['host_response_time'])
# Create Dummy Variables forneighbourhood_group_cleansed variable without Prefix parameter
ngclean = pd.get_dummies(dframe['neighbourhood_group_cleansed'])
# Create Dummy Variables for property_type variable without Prefix parameter
ptype = pd.get_dummies(dframe['property_type'])
# Create Dummy Variables for bed_type variable
bed = pd.get_dummies(dframe['bed_type'])
# Create Dummy Variables for cancellation_policy variable
cpolicy = pd.get_dummies(dframe['cancellation_policy'])

# Concatenate All the Dummy Variables inot the Dataframe
dum_df = pd.concat([dframe,hrtime,ngclean,ptype,bed,cpolicy], axis=1)
dum_df.info()
```

```python
# Create Dummy Variables for Amenties variable
amen = dframe['amenities'].str.get_dummies(sep=',')
dum_df = pd.concat([dum_df,amen], axis=1)
dum_df.columns
```
